### Проверка расчетов

Для начала проверим правильно ли рассчитана эффективность обогащения.

Она расчитывается по формуле:

$$Recovery=\frac{C(F-T)}{F(C-T)}*100\%$$

Где:
- C — доля золота в концентрате после флотации/очистки
- F — доля золота в сырье/концентрате до флотации/очистки
- T — доля золота в отвальных хвостах после флотации/очистки

```python
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

train['recovery_calc'] = ((C*(F-T))/
                          (F*(C-T)))*100

recovery_mae = mae(train['recovery_calc'],train['rougher.output.recovery'])
recovery_mae
```

Полученное значение MAE очень близко к 0, это говорит о том, что эффективность обогащения посчитана правильно

```python
def loses (df):
    len_loses = np.sum(df.isna(), axis=1)!=0
    len_loses = len_loses.sum()
    
    len_data = df.shape[0]
    
    loses_frac = len_loses / len_data
    
    print(f'В выборке {len_loses} строк с пропусками')
    print(f'Это составляет {loses_frac:.2%} от всей выборки')
```

### Функции для расчета sMAPE

Зададим функции для расчетов sMAPE и final sMAPE

```python
def smape (true, pred):
    result = (np.abs(true-pred)/
             ((np.abs(true) + np.abs(pred))/2))
    return result.mean() * 100 
```

```python
def final_smape (true, pred):
    smape_rougher = smape (true.values[:,0], pred[:,0])
    smape_final = smape (true.values[:,1], pred[:,1])
    result = .25*smape_rougher + .75*smape_final
    return result
```

```python
smape_score = make_scorer(final_smape,  greater_is_better = False)
```
