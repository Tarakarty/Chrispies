## Предобработка данных

### Изменение названия столбца
Начнем с простого: изменим название столбца `cityCenters_nearest`

```python
df = df.rename({'cityCenters_nearest':'city_centers_nearest'}, axis=1)
```

### Дубликаты

Явных дубликатов датасет не содержит, поэтому займемся неявными дубликатами в названиях населенных пунктов.
- Избавимся от разницы в регистрах букв.
- Избавимся от типов населенных пунктов: деревня, село, пгт и т.д.

```python
new_locals = (df['locality_name']
 .str.lower()
 .str.replace('ё','е')
 .str.replace('поселок ','')
 .str.replace('городского типа ','')
 .str.replace('городской ','')
 .str.replace('деревня ','')
 .str.replace('платформа ','')
 .str.replace('село ','')
 .str.replace('садовое товарищество ','')
 .str.replace('садоводческое некоммерческое товарищество ','')
 .str.replace('при железнодорожной станции ',''))
new_locals.sort_values().unique()
```

```python
df.loc[:,'locality_name'] = new_locals
```

### Пропуски

#### Пропуски, которые мы пропустим

Некоторые пропуски невозможно заполнить исходя из имеющихся данных, заполнение же других не принесет пользы исследованию.
1) Время публикации
 - Самым логичным кажется предположение, что пропуски соответствуют актуальным объявлениям. Т.е. квартиры все еще продаются. Т.к. неизвестна даже дата снятия данных. Максимум, что мы можем сделать это рассчитать самую позднюю дату снятия объявления и предложить ее для рассчета во всех пропусках, но это может привести к появлению слишком "быстрых" квартир, это исказит данные. Исходя из всего этого пропуски оставим, а в голове будем держать факт наличия активных объявлений и радоваться актуальности данных)
 
2) Пруды и парки
 - Количество пропусков сопоставимо по обеим категориям (количеству и расстоянию) причем расстояние не указывали гораздо чаще, чем количество. Поэтому можно предположить, что а) заполнявшие анкеты не указывали данные т.к. считали, что это не поможет быстрее и выгоднее продать квартиру и б) для небольших населенных пунктов пункт анкеты отсутствует в принципе, за ненадобностью. Оставим как есть.
 
3) Аэропорт и центр
 - Количество пропусков почти совпадает с пропусками в расстояниях до водоемов и парков, что говорит скорее о справедливости гипотезы б) из прошлого пункта. Пусть останется неизменным.
 
4) Название НП
 - Можно провести расследование и сопоставить информацию по расстояниям до аэропорта, парка, водоемам... Наложить это все на карту и найти наиболее подходящие НП. Но это задача для детектива из сериала, мы тут другими вещами занимаемся)

5) "Квартиы-неквартиры"
 - 0-комнатные квартиры оставим, возможно так владельцы решили указать, что это квартиры-студии

#### Площади
##### Студии

Для начала займемся заполнением площадей в студиях.

```python
df_studio = df[df['studio'] == True].copy()#.reset_index(drop=True)
df_studio.info()
```

В заполнении нам поможет определение: "*Квартира-студия – это такая жилая площадь, в которой нет перегородок*". Т.е. из жилой площади можно не выделять площадь кухни.
Исходя из этой предпосылки заполним жилую площадь по среднему или медиане отношения общей площади к жилой. Площадь кухни установим в 5 метров (минимальная разрешенная законом площадь кухни) сразу для всех пропусков в этих площадях.

```python
(df_studio['living_area'] / df_studio['total_area']).agg(['mean','median'])
```

Разница указывает, что есть студии с большими коридором и ванной. J

Для заполнения возьмем медиану.

```python
df.loc[(df['studio']==True) & (df['living_area'].isna()),'living_area'] = df.loc[(df['studio']==True) & (df['living_area'].isna()),'total_area'] * 0.689405

df[df['studio']==True]['living_area'].isna().sum()
```

```python
df.loc[(df['studio']==True) & (df['kitchen_area'].isna()),'kitchen_area'] = 5
df[df['studio']==True]['kitchen_area'].isna().sum()
```

##### Остальные

По аналогии со студиями заполним жилую площадь для остальных квартир.

```python
df_nostudio = df[df['studio'] == False].copy()
(df_nostudio['living_area'] / df_nostudio['total_area']).agg(['mean','median'])
```

```python
df.loc[(df['studio']==False) & (df['living_area'].isna()),'living_area'] = df.loc[(df['studio']==False) & (df['living_area'].isna()),'total_area'] * 0.567164

df['living_area'].isna().sum()
```

#### Потолки (+ аномалии)

Для начала посмотрим характер распределения значений

```python
df['ceiling_height'].describe()
```

Согласно открытым данным высота потолков должна быть не менее 2,5 метров и в среднем находится в интервале 2,5-3,2 метра.
Это согласуется со значениями в датасете, поэтму пустые значения заполним медианой.

```python
df['ceiling_height'].fillna(df['ceiling_height'].median(), inplace=True)
df['ceiling_height'].isna().sum()
```

Теперь займемся аномалиями.

В данных есть значения превышающие даже 20 метров. Посмотрим на самый выбивающийся пример с потолками, размером с высотку.

```python
df[df['ceiling_height']>50]
```

Объявление выглядит неправдоподобно. 5 балконов на 25 метров площади, без единой фотографии. Или содержит много ошибок.
Учитывая что такое объявление всего одно проще его удалить, чем пытаться установить истину.

```python
df.drop(index=22869, inplace=True)
df.reset_index(drop=True,inplace=True)
df[df['ceiling_height']>50]
```

Потолки 20м и выше поделим на 10. Очень похоже, что это просто опечатки

```python
df.loc[df['ceiling_height']>=20,'ceiling_height'] = df.loc[df['ceiling_height']>20,'ceiling_height'] / 10
```

```python
df[df['ceiling_height']>=5]
```

Таким образом осталось всего 14 объявлений с неправдоподобно высокими потолками. Их немного, как логично эти значения заменить непонятно. Их исключаем тоже. За порог возьмем 5 метров.

```python
df = df[df['ceiling_height']<5]
```

#### Этажи

### Категоризация этажей

Для заполнения этого столбца создадим функцию, возвращающую значения: первый, последний, другой.

```python
def floor_category (row):
    if row['floor'] == 1:
        return 'first'
    elif row['floor'] == row['floors_total']:
        return 'last'
    else:
        return 'other'
```

И применим эту функцию к датафрейму

```python
df['floor_category'] = df.apply(floor_category, axis=1)
df[['floor','floors_total','floor_category']].head()
```
